```python
import pandas as pd
```

```python
df = pd.read_csv("Flavors.csv")
# df.index += 1
# df.index.name = "ID"
df
```

```python
group_by_frame = df.groupby("Base Flavor")[["Flavor Rating", "Texture Rating", "Total Rating"]]
```

```python
group_by_frame.mean() 
# mean can only be applied on numeric columns that is why we selected only those columns in the above line 
# if we didnt select only those columns it would have thrown an error
```

```python
# df.groupby("Base Flavor")[["Flavor Rating", "Texture Rating", "Total Rating"]].mean()
```

```python
df.groupby("Base Flavor").count()
```

```python
df.groupby("Base Flavor")[["Flavor Rating", "Texture Rating", "Total Rating"]].min()
```

```python
df.groupby("Base Flavor").max()
```

```python
df.groupby("Base Flavor")[["Flavor Rating", "Texture Rating", "Total Rating"]].sum()
```

```python
# it is neccessary to select the column for aggregation and also pass a dictionary with column name as key and list of aggregation functions as value
df.groupby("Base Flavor").agg({"Flavor Rating": ["min", "max", "mean", "sum", "count"]})
```

```python
df.groupby("Base Flavor").agg({"Texture Rating": ["min", "max", "mean", "sum", "count"]})
```

```python
# aggregation on multiple columns
df.groupby("Base Flavor").agg({"Flavor Rating": ["min", "max", "mean", "sum", "count"], "Texture Rating": ["min", "max", "mean", "sum", "count"]})
```

```python
# we can group by multiple columns as well
group_by_frame = df.groupby(["Base Flavor", "Liked"])[["Flavor Rating", "Texture Rating", "Total Rating"]].mean()
```

```python
# as well as grouping but we can aggregate on multiple columns as well
df.groupby(['Base Flavor','Liked']).agg({'Flavor Rating': ['mean','max','count','sum']})
```
