```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("world_population.csv")
df
```

```python
pd.set_option("display.float_format", lambda x: "%.2f" % x)
df
```

```python
df.info() # to get a concise summary of the DataFrame
```

```python
df.describe() # to get statistical summary of numerical columns
```

```python
df.isnull().sum() # to check for missing values in each column
```

```python
df.nunique() # to get unique values in each column
```

```python
df.sort_values(by="2022 Population").head(10) # to get top 10 countries by population in 2022 in ascending order
```

```python
df.sort_values(by="2022 Population", ascending=False).head(10) # to get top 10 countries by population in 2022 in descending order
```

```python
df.corr(numeric_only=True) # to get correlation matrix of numerical columns
# used numeric_only=True to avoid warnings about non-numeric data
# this will help us understand the relationships between different numerical variables in the dataset
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
# used annot=True to display correlation values on the heatmap
plt.rcParams["figure.figsize"] = (20, 7)
plt.show()
```

```python
# in the figure the same colors represent high correlation and different colors represent low correlation
```
